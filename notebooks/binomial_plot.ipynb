# Calculation of feasible Event Logs for SHApley experiments

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

def S(n, k):
    return sum(comb(n, i) * k**i for i in range(1, n+1))#*5/12.49/60#Number of necessary ELs #*5 Miners

# Define range of n values
n_values = np.arange(1, 3+1) # Number of features
k_values = [2,3,4]  # Number of values per feature

plt.figure(figsize=(10, 6))
for k in k_values:
    y_values = [S(n, k) for n in n_values]
    plt.plot(n_values, y_values, label=f'k={k}')

plt.xlabel('Number of features')
plt.ylabel('Number of necessary feasible event logs for SHAp experiment')
plt.title('Plot of S(n, k) for different k values')
plt.legend()
plt.grid()
plt.show()
```
